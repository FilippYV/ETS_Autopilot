```python
wheel_ets_train_x = pd.DataFrame(columns=[i for i in range(1, 64*64 + 1)])
```

```python
size = 512
combined_mask = np.zeros((size, size))
model = YOLO(get_path_weight_model('best_v2_n.pt'))

for file in png_files:
    results = model(file, imgsz=512, conf=0.65, show=False, device='cuda')
    
    combined_mask -= combined_mask * 0.34
    
    if results[0].masks is not None:
        for i in results[0].masks.data:
            combined_mask += i.cpu().numpy()
    new_index = len(wheel_ets_train_x)

    # Добавьте новую строку в DataFrame с использованием метода loc
    name = file.split(f'\\')[-1].replace('.png', '')
    wheel_ets_train_x.loc[name] = np.resize(combined_mask, (64,64)).flatten()
```
